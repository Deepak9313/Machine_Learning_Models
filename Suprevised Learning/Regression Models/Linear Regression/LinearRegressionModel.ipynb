```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
```

Import Dataset

```python
data = pd.read_csv('Salary_Data.csv',header=None)
data.columns = ['YearsExperience', 'Salary']
data = data[1:]
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 100)
data
```

```python
x = np.asarray(data['YearsExperience'].values.reshape(-1,1))
y = np.asarray(data["Salary"].values.reshape(-1,1))
```

Apply the train test function

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

Create Model of LinearRegression

```python
model = LinearRegression()
```

Fit The Model

```python
model.fit(x_train,y_train)
```

Now Let's Find The Prediction

```python
y_pred = model.predict(x_test)
```

Now We have two value 
y_predict => Predicted values
y_test => Test Values
Now We can see the Difference and See the Regression Line

Lets Check The Score of Model Through
 :- mean_squared_error(y_test,y_pred)
 :- r2_score(y_test,y_pred)

```python
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)
```

A Low Mse is good
If your target variable values range from 1 to 10, an MSE of 50 is terrible.
If your target variable values range from 10,000 to 100,000, an MSE of 50 is excellent.
The Value of R2 lies Between 0 to 1 
:- close to 1 give you accurate results

```python
print(f'Mean Squared Error: {mse}')
print(f'R2 Score: {r2}')
```

Now Lets Check Coefficient and Intercept

```python
print(f'Intercept: {model.intercept_}')
print(f'Coefficient: {model.coef_}')
```
