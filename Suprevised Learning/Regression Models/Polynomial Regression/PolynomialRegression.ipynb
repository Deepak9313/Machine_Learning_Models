```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.metrics import r2_score,mean_squared_error
```

```python
np.random.seed(43)
x = np.linspace(-10,10,30)
y = 3 * x**2 + 2 * x + 1 + np.random.randn(30) * 10

plt.scatter(x, y)
plt.show()
```

As If we have the data above we need to fit the degree but how you find the right one
a high degree give you a goood r2 score but it risks overfitting
overfitting: means when data is too less for applying the model
Recommended Approach:
Degree 1 (Linear): If the data is roughly linear.
Degree 2 or 3 (Quadratic/Cubic): Captures curvature without overfitting.
Degree 5-10: May fit better but risks overfitting.
Degree 150: Almost always overfitting.

```python
mymodel = np.poly1d(np.polyfit(x, y, 2))
y_pred = mymodel(x)
r2 = r2_score(y, y_pred)
mse = mean_squared_error(y, y_pred)
print(f"R2: {r2}")
print(f"MSE: {mse}")
```

```python
plt.scatter(x,y)
plt.plot(x, y_pred, color='red')
plt.show()
```
